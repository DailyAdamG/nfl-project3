```python
from sklearn import tree
import pandas as pd
import os
```

```python
QB_data = pd.read_csv("../nfl-project3/QB Machine Learning.csv")
```

```python
QB_data = QB_data.loc[QB_data["Touches"] > 0]

QB_data
```

```python
target = QB_data["RookieAV"]
```

```python
data = QB_data.drop(["Year", "Name", "College", "POS", "SixtyYdShuttle", "BenchPress", "ThreeCone", "BroadJump", "VertLeap", "Shuttle", "RookieAV"], axis=1)
data.fillna(data.mean(), inplace=True)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=25)
```

```python
clf = tree.DecisionTreeRegressor()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=10000)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
NewQBs = pd.read_csv("../nfl-project3/Quarterbacks 2017.csv")

NewQBs.fillna(data.mean(), inplace=True)

predictionDF = NewQBs.drop(["Year", "Name", "College", "POS"], axis = 1)

yhat = rf.fit(X_train, y_train).predict(predictionDF)
```
